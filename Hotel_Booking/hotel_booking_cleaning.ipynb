```python
import pandas as pd
```

```python
hotel=pd.read_csv(r'D:\Data Analytics\Python\Panda\Data Cleaning\hotel_bookings_messy.csv')
```

```python
# first lookat the dataset
hotel
```

```python
# drop/rename column
hotel.columns
```

```python
hotel.drop(columns=['Unnamed: 0'],inplace=True) #droping

# here axis=1 because we are telling panda that we are renaming column not
# hotel.rename({'adults':'num_adults','children':'num_children','babies':'num_babies'},inplace=True,axis=1)

# or 
hotel.rename(columns={'adults':'num_adults','children':'num_children','babies':'num_babies'},inplace=True)
hotel.columns
```

```python
# Nans 
# below both return null or not available values in column 
# hotel.isna().sum()
# hotel.isnull().sum()

# here from this syntax we found out percentage of null in column
hotel.isnull().sum()*100/len(hotel)
```

```python
# since company column has 94% null we will drop it 
# and agent has 14% null , we will not drop row with null instead we will replace null with placeholder value ex- -1 , here data is numeric so -1
# and also children have null 
```

```python
# this returns unique values
# hotel['agent'].unique()

# here am filling -1 to all na in agents column 
hotel['agent'].fillna(-1,inplace=True)
```

```python
# this return all row with children null
# hotel[hotel['num_children'].isna()]

# since there are 4 row with null in children we'll drop it 
hotel.dropna(subset='num_children',inplace=True)
```

```python
# hotel[hotel['country'].isna()][['hotel','country']]
# here i filled unknown to all null value in country column
hotel['country'].fillna('Unknown',inplace=True)
```

```python
# since company has 94% of null we will drop it
hotel.drop(columns=['company'],inplace=True)
```

```python
# check column datatype
# this is how we change the data type of the columns
hotel=hotel.astype({'is_canceled':'boolean','is_repeated_guest':'boolean','num_children':'int','agent':'int'})
hotel.dtypes
```

```python
# Bin Columns
# Grouping different columns if they need to be
# Binning them , grouping them into a bucket like we do to age 0-10,11-20 like that 
# we will bin the lead_time
# hotel['lead_time'].unique()
# hotel['lead_time'].describe()

bins= [0,100,200,300,400,500,600,700,800]
labels=['0-100','101-200','201-300','301-400','401-500','501-600','601-700','701-800']

hotel['lead_time_bin']=pd.cut(hotel['lead_time'],bins=bins,labels=labels)

hotel[['lead_time','lead_time_bin']]
```

```python
# seperate data 
# seperate data from one column into multiple column

# here am saying split on basis of - and i want [0] index part since it return both part from them i want column which is at [0] index
hotel['arrival_date_month']=hotel['arrival_date'].str.split('-',expand=True)[0]
hotel['arrival_date_year']=hotel['arrival_date'].str.split('-',expand=True)[1]


# moving column to specific position
# pop() - returns item and drop from frame
col_to_move=hotel.pop('arrival_date_month')
hotel.insert(4,'arrival_date_month',col_to_move)


col_to_move=hotel.pop('arrival_date_year')
hotel.insert(5,'arrival_date_year',col_to_move)

hotel.head()
```

```python
# here i droped arriavl_Date column which is now of now use.
hotel.drop(columns=['arrival_date'],inplace=True)
```

```python
# String Cleaning
# here we will use regex
# what i did is i replace every instance of *,n,^ , to write it i use \ and wrote them and replace them with empty string make sure to regex is true. 
hotel['hotel']=hotel['hotel'].replace(r'[\*\n\^]','',regex=True)
hotel['hotel'].unique()
```

```python
# Removing Duplicate
# keep = first/last/false

# keep='first' (default): Marks all duplicates as True, except the first occurrence, which is marked as False. 

# keep='last': Marks all duplicates as True, except the last occurrence, which is marked as False. 

# keep=False: Marks all occurrences of duplicates as True. 
# hotel.duplicated(keep=False)


# this print all duplicate rows
# hotel.loc[hotel.duplicated(keep=False)]

# subset specify what columns to look, if not give provided it drop duplicate from all columns
# hotel.drop_duplicates(subset=)
# below keep=first specifies that keep first instance of the record and delete the duplicate 
hotel.drop_duplicates(keep='first',inplace=True)
```
