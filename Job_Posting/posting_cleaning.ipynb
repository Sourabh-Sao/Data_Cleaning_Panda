```python
import pandas as pd
```

```python
posting=pd.read_csv(r"D:\Data Analytics\Python\Panda\Data Cleaning\postings.csv")
```

```python
# first step making a copy of original
posting_cleaned=posting.copy()
# posting_cleaned.head()
posting_cleaned.columns
```

```python
# correcting column name to be more descriptive
posting_cleaned.rename(columns={'title':'job_title','description':'job_description','applies':'num_applicant'},inplace=True)
```

```python
# Cleaning column datatype
# posting_cleaned.dtypes

# posting_cleaned['zip_code']=posting_cleaned['zip_code'].astype("str")
# posting_cleaned['fips']=posting_cleaned['fips'].astype("str")

# or
# posting_cleaned=posting_cleaned.astype({'zip_code':'str','fips':'str'})
```

```python
# droping columns that are of no use
posting_cleaned.drop(columns={'med_salary','closed_time','skills_desc'},inplace=True)
```

```python
# String cleaning in columns containing string
# text standardization, removing leading or trailing spaces or either making string all uppercase or lowercase

# removing white space
# strip() - removes leading or trailing white space
posting_cleaned['job_title']=posting_cleaned['job_title'].str.strip()
posting_cleaned['job_title']=posting_cleaned['job_title'].str.lower()
```

```python
# handling missing values
# posting_cleaned.isna().sum()

# here i fill na with 0 and corrected the column datatype from boolean to int 
# posting_cleaned['remote_allowed'].fillna(0,inplace=True)
posting_cleaned['remote_allowed']=posting_cleaned['remote_allowed'].astype('int')
```

```python
# droped all rows where company_name is null/empty
posting_cleaned.dropna(subset=['company_name'],axis=0,inplace=True)
```
